```python
import numpy as np
import pandas as pd
import html5lib
print ('Libraries imported.')
```

```python
page='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
table= pd.read_html(page)
print('Extracted table')
```

```python
table[0].head()
```

```python
data=table[0]
data
```

```python
index=data[data['Borough']=='Not assigned'].index
data.drop(index,inplace=True)
data.reset_index(drop=True, inplace=True)
print(data.shape)
```

```python
data.head()
```

### In the data cleaning process I need not to group the two neighbourhoods under one postal code. Because it was sorted in the wikipadia.

```python
data_loc=pd.read_csv('http://cocl.us/Geospatial_data')
data_loc.head()
```

```python
data_Toronto=pd.merge(data,data_loc)
data_Toronto.head()
```
